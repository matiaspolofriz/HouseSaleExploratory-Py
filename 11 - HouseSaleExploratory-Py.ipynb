## 4.1.- Matriz de Correlación

###### armo una matriz de correlación (mapa de calor) para ver cuales son las variables más correlacionadas

###### declaro una variable corrmat = df_train.corr() con todas las correlaciones entre variables

```python
corrmat = df_train.corr(numeric_only = [False])
```

###### f,ax para declarar un plt.subplots
###### Y finalmente hacemos el sns.heatmap

```python
f,ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=8, square = True)
```

## 4.2.- Matriz de Correlación con Números

###### lo mismo pero ahora con números - para ver quien se correlaciona más - sin llegar al punto que las 2 variables sean basicamente lo mismo claro

###### declaro k=10

```python
k=10
```

###### Luego cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index

```python
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
```

```python
cols
```

###### Luego genero esta variable con los coeficientes de coorelacion cm = np.corrcoef(df_train[cols].values.T)

```python
cm = np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

###### sns.set(font_scale=1.25)

```python
sns.set(font_scale=1.25)
```

###### Armo el heatmap con hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)

```python
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

### Podemos ver que:

- Calidad, AreaVivienda y area Sotano estan muy fuertemente correlacionadas con PrecioVenta
- Las variables de garage están super correlacionadas
- Sotano y 1er piso son basicamente hermanos gemelos
- Y AñoConstruccion esta ligeramente correlacionada con PrecioVenta, parece que en un futuro habra que hacer un analisis de serie de tiempo más serio.
